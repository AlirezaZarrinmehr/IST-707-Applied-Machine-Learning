```python
import pandas as pd

import sklearn.model_selection

from sklearn.model_selection import train_test_split

df = pd.read_csv("/Users/zarrinmehr/Desktop/11- IST.707.M002.FALL22.Applied Machine Learning 16797.1231/week 5/HW 5/HW4-data-fedPapers85.csv")

df.head(12)
```

```python
dfDispt = df[df.author == 'dispt']
dfDispt
```

```python
dfOther = df.drop(range(0,11))
dfOther
```

```python
training_data, testing_data = train_test_split(dfOther, test_size=0.2, random_state=25)

print(f"No. of training examples: {training_data.shape[0]}")
print(f"No. of testing examples: {testing_data.shape[0]}")
```

```python
testing_data_Dispt = pd.concat([testing_data, dfDispt])
testing_data_Dispt
```
